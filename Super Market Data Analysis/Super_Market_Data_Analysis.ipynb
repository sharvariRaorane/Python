```python
#Which location has the highest and lowest sales?
city_data = data.groupby('Location')
locations = data['Location'].unique()
location_sales = city_data['Total'].sum()
plt.bar(locations,location_sales)
```

```python
#Represent the sales on a bar chart, also show the market share for each location using pie chart
plt.pie(location_sales,labels=locations,autopct = '%1.1f%%')
plt.show()
```

```python
#which location has more females customers and which locations has more male
city_gen_data = data.groupby(['Gender','Location']).count()['Total']
unstack_gen_city = city_gen_data.unstack(level=0)
unstack_gen_city.plot(kind='bar')
```

```python
#which branch has more member vs which branch has less members
loc_mem_data = data.groupby(['Location','Member']).count()['Total']
unstack_loc_mem_data = loc_mem_data.unstack()
unstack_loc_mem_data.plot(kind='bar')
```

```python
#Which branch has highest rating and which branch has lowest
data.groupby('Location')['Rating'].mean().plot(kind='bar')
```

```python
#Which city has more female shopping
data_shopper = data.groupby(['City','Gender']).count()['Total'].unstack().plot(kind='bar')
```

```python
#who spend more men or women
data.groupby('Gender').sum()['Total'].plot(kind='bar')
```

```python
#Which type of customer spend more member or non member
data.groupby('Member').sum()['Total'].plot(kind='bar')
```

```python
#which productline sells more
data.groupby('Category').count()['Total'].plot(kind='bar')
```

```python
#which productline is popular among men and women
data.groupby(['Category','Gender']).count()['Total'].unstack().plot(kind='bar')
```

```python
#what days of the month make most sales
data['Day'] = pd.to_datetime(data['Date']).dt.day
data.groupby('Day').sum()['Total'].plot()
```

```python
#which month makes the most sales
data['Month'] = pd.to_datetime(data['Date']).dt.month
data.groupby('Month').sum()['Total'].plot(kind='bar')
```

```python
#Which hour makes most sales
data['Hour'] = pd.to_datetime(data['Time']).dt.hour
data.groupby('Hour').sum()['Total'].plot()
```

```python
#What time customers make cash payments vs epayments
data.groupby(['Hour','Payment']).count()['Total'].unstack().plot(kind='bar')
```
