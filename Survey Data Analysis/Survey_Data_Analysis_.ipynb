Data Visualization

```python
from matplotlib import pyplot as plt
```

```python
ed_percent = data['EdLevel'].value_counts(normalize = True)*100
ed_percent = ed_percent.sort_values()
ed_percent.plot(kind='barh')
plt.title("Distribution of Education")
plt.xlabel("Percentage")
plt.ylabel("Education Level")
plt.show()
```

```python
learn_methods = data['LearnCode'].str.split(";").explode()
```

```python
method_count = learn_methods.value_counts()
```

```python
method_percentage = (method_count/89184)*100
```

```python
method_percentage
```

```python
method_percentage = method_percentage.sort_values()
method_percentage.plot(kind='barh')
plt.title("Learning Methods")
plt.xlabel("Percentage")
plt.ylabel("Learn Code")
for index,value in enumerate(method_percentage):
  plt.text(value,index,f'{value: .2f}%',va='center')
plt.show()
```
